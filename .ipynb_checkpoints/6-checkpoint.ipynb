# 6.기억장치 관리기법(주,가상)
#### 기억장치 관리 전략
반입전략
배치전략:최초적합/최적적합/최악적합
#### 단편화 해결 방법
통합기법/집약기법,압축,쓰레기수집
#### 단일분할할당>스와핑
필요에 따라 다른 프로그램과 교체
#### 단일분할할당>오버레이
작업의 크기가 커서 사용자 기억 공간에 수용될 수 없을때 필요한 부분만 교체
#### 단일분할할당>고정분할
주기억장치를 미리 몇개의 고정된 개수와 크기의 부분으로 분할하여 여러 개의 프로그램이 동시에 적재되어 실행되게 하는 방법
#### 단일분할할당>가변분할
고정분할할당기법의 단편화를 줄이기 위한 것
#### 가상기억장치 개요
일부분을 주기억장치로 사용
#### 페이징 기법
가상기억장치 보관되어 있는 프로그램과 주기억장치 영역을 동일한 크기로 나눈후 나눠진 프로그램을 동일하게 나눠진 주기억장치의 영역 페이지 프레임에 적재시켜 실행하는 기법
주소변환/페이지부재
#### 세그멘테이션
가상기억장치에 보관되어 있는 프로그램을 논리적단위로 나눈후 주기억장치에 적재시켜 실행(메모리절약)
#### 가상기억장치의 성능에 영향을 미치는 요인
워킹셋
스래싱
구역성

# 7.페이지 교체 알고리즘
#### OPT최적교체
앞으로 가장 오랫동안 사용하지 않을 페이지를 교체 - 실현가능성 없음
#### LRU
가장 오랫동안 사용하지 않은 페이지 교체
#### LFU
사용횟수가 가장 적은 페이지 교체
#### NUR
최근에 사용하지 않은 페이지 교체

# 9. 파일관리
#### 파일시스템 특징
소멸성/활성률/크기
#### 순차파일
기억장치의 효율이 높다
검색시에 효율이 나쁘다
일괄처리 적합
#### 색인 순차 파일
실린더 색인,트랙색인,마스터색인
#### 직접파일
해싱등의 사상함수를 사용하여 레코드 키에 의한 주소 계산을 통해 레코드 접근할 수 있도록 구성한 파일
#### 파일디스크립터(파일제어블록)
파일을 관리하기 위한 시스템이 필요로 하는 파일에 대한 정보를 갖는 제어 블록
#### 디렉토리 구조
1단계구조:가장간단,관리불편
2단계구조:마스터/사용자파일디렉토리
트리구조:루트/종속(서브)디렉토리
비순환 그래프 구조
부 디렉토리,파일공유,사이클x
일반그래프구조:원하는 파일로 접근이 쉽다
#### 디스크 공간 할당 기법
연속할당
불연속할당
#### 자원 보호 기법
접근제어행렬
접근제어리스트
권한리스트
#### 파일 보호 기법
파일의 명명
비밀번호
접근제어
#### 보안 기법
외부보안
내부보안
사용자 인터페이스 보안
#### 암호화 기법
비밀키시스템: 암호화키=복호화키
공용키시스템: 암호화키<>복호화키
