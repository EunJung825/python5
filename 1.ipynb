# .데이터 베이스

## 데이터 베이스 정의
특정 조직이 업무 수행하는데 필요한 관련성 있는 자료들의 집합체(통합,저장,운영,공용)
## 데이터베이스 시스템 도입배경: 파일 시스템의 문제점해결
데이터 중복성과 데이터 종속성 발생:데이터 무결성 위해 가능성 높음
## 데이터베이스 시스템 개념
물리적데이터 독립성
논리적데이터 독립성
## 데이터베이스 특징
실시간접근/계속적인변화/공용/내용에의한참조
## 데이터베이스 시스템 구성
정보시스템/자료처리시스템(일괄처리,온라인실시간처리,분산처리)
## 데이터웨어하우스
의사결정자원시스템지원하는 주체적,통합적,시간적 데이터의 집합체

# 2 DB 2강-DBMS, 스키마, 데이터언어, DBA
##  DBMS 정의
사용자의 요구에 따라 DB생성, 관리해 주는 S/W
##  DBMS 필수 기능
정의,조적,제어
## DBMS(DB) 장점
독립성,중복없어짐,공동이용,일관성,무결성,표준화,통합,최신의데이터유지,실시간처리
## DBMS(DB) 단점
전문가 부족, 비용증가, 시스템복잡
## 스키마
데이터베이스의 구조와 제약조건에 대한 명세를 기술한 것
## 스키마 특징
데이터사전,구조정의,메타데이터
## 스키마3계층
외부스키마,개념스키마,내부스키마
## 데이터언어
DBMS의 3가지 기능을 구현하기 위한 언어
## 데이터언어종류
정의(DDL)/조작어(DML)/제어어(DCL)
## 데이터베이스 사용자
응용프로그래머,일반사용자,DB관리자
## DBA의 역할
데이터베이스 시스템의 모든 관리와 운영에 대한 책임을 지고 있는 사람이나 그룹으로서 DDL,DCL을 사용

# 데이터베이스 설계
현실세계의 업무적인 프로세스를 컴퓨터 세계로 데이터베이스화 하기 위한 과정
## 데이터베이스 설계 순서
개념. 논리. 물리
## 요구조건분석
## 개념적 설계
개체 타입과 이들 간의 관계 타입을 이용해 현실 세계를 개념적으로 표현
개체/속성/관계
## 논리적 설계
목표DBMS에 맞추어 논리적 모델로 설계
## 물리적 설계
저장레코드 양식의 설계 및 물리적 구조 데이터 표현
## 구현
목표 DBMS DDL로 스키아를 작성, 응용 프로그램을 위한 트랜잭션을 작성하는 관계
## 데이터 모델 정의
현실 세계의 데이터 구조를 컴퓨터 세계의 데이터 구조로 기술하는 개념적 도구
## 데이터 모델 종류
개념적/논리적
## 데이터 모델 구성요소
구조,연산,제약조건
## 데이터 모델 용어
ER모델
## 관계의 종류(관계타입)
1:1, 1:N, N:M

# -논리적데이터모델, 관계데이터모델
## 논리적 데이터 모델의 종류
관계형,계층형, 네트워크형
## 요구분석->개념적설계=>논리적설계
## 관계형 데이터베이스의 릴레이션 구조
도메인,차수,카디날리티,널(해당없음)
## 릴레이션 특징
1) 한 릴레이션에 정의된 튜플들은 모두 다르다. 
2) 한 릴레이션에 정의된 튜플들은 순서에 무관하다. 
3) 튜플들은 시간에 따라 변한다. 
4) 릴레이션 스키마를 구성하는 속성들도 순서에 무관하다. 
5) 속성의 명칭은 유일해야 하지만, 속성의 값은 동일해도 된다. 
6) 속성은 더 이상 쪼갤 수 없는 원자값으로 구성된다. (속성 값은 분해 X, 다중값 X)
7) 릴레이션을 구성하는 튜플을 유일하게 식별하기 위한 속성들의 부분집합을 키(Key)로 설정한다. 
## 키
슈퍼키/후보키/기본키/대체키/외래키
## 제약조건(무결성)
참조무결성/개체무결성/도메인무결성

# 정규화, 관계데이터연산, 관계대수
## 관계형데이터베이스 모델링

중복방지,효율적으로 저장
## 제1정규형
모든 속성이 원자 도메인 만으로 되어 있는 정규형
## 이상

갱신이상/삽입이상/삭제이상
## 제2정규형
제1정규형이고, 부분함수적 종속을 제거하여 완전(충분한) 함수적 종속을 만족하는 정규형
함수적 종속
## 제3정규형
제2정규형, 이행적 함수적 종속 관계 제거하여 비이행적 함수적 종속 관계 만족
## BCNF
제 3정규형이고, 결정자가 후보키가 아닌 함수 종속 제거 모든 결정자가 후보키이어야 한다
## 제4정규형
다치종속제거
## 제5정규형
조인 종속성 이용

## 정규화 특징

## Division
나누어지는 릴레이션인 A는 릴레이션 B의 모든 내용을 포함한 것이 결과 릴레이션이 된다
## 합집합
릴레이션 A 또는 B에 속하는 튜플들로 구성된 릴레이션이다
## 교집합
 릴레이션 A 와 B에 공통적으로 속하는 튜플들로 구성된 릴레이션이다
 ## 차집합
 릴레이션 A에만 있고 B에는 없는 튜플들로 구성된 릴레이션이다
 ## 카티션 프로덕트
